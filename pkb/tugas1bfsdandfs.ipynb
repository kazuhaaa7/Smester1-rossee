```python
queue = [] 
visited =  []

def bfs(visited, graph, node):  
     visited.append(node) 
     queue.append(node) 
     while queue:       
            m = queue.pop(0) 
            print (m, end = " ") 
            for neighbour in graph[m]:    
                if neighbour not in visited: 
                    visited.append(neighbour) 
                    queue.append(neighbour)                         
print("Urutan node yang dikunjungi :") 
bfs(visited, graph, '2A') 
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Graph sesuai dictionary
graph = { 
    '2A': ['5B', '2C'],   
    '5B': ['4D', '1E'],
    '2C': ['0F','1G'],
    '4D': ['0H'],
    '0H': [],
    '1E': ['1I'], 
    '1I': [],
    '0F': ['0J'],
    '0J': [],
    '1G': ['9K'],
    '9K': ['0L'],
    '0L': []
}

# Buat edges dari dictionary
edges = []
for node, neighbors in graph.items():
    for neighbor in neighbors:
        edges.append((node, neighbor))

# Buat directed graph
G = nx.DiGraph()
G.add_edges_from(edges)

# Atur posisi manual biar horizontal bercabang
pos = {
    '2A': (0, 0),
    '5B': (1, 1),
    '2C': (1, -1),
    '4D': (2, 2),
    '1E': (2, 0.5),
    '0F': (2, -0.5),
    '1G': (2, -1.5),
    '0H': (3, 2),
    '1I': (3, 0.5),
    '0J': (3, -0.5),
    '9K': (3, -1.5),
    '0L': (4, -1.5),
}

# Gambar graph
plt.figure(figsize=(10, 6))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=2000,
    node_color="black",
    font_color="white",
    font_size=10,
    edge_color="gray",
    arrows=False
)

plt.title("Graph NIM 252410101090", fontsize=12)
plt.show()
```
