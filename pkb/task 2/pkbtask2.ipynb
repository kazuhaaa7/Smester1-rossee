```python
# imporr library yg diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# membuat data

data = {
    "Jumlah_Iklan" : [10, 15, 12, 20, 25, 18, 22, 8, 28, 16],
    "Jumlah_Pembeli" : [150, 160, 155, 180, 200, 175, 190, 140, 210, 170]
}

datas = data
df = pd.DataFrame(datas)

df
```

```python
#ekspolrasi data(memviusualisasikan data) 
sns.scatterplot(data=df, x = "Jumlah_Iklan", y = "Jumlah_Pembeli")
plt.title('Jumlah iklan dengan jumlah pembeli')
plt.show()
```

```python
# persiapan data
x = df[['Jumlah_Iklan']] #features/input,independent, apabila input memanglah harus memakai 2 kurung siku- hasilnya df (machine learning) 
y = df['Jumlah_Pembeli'] #target/output,dependent, ga harus memakai 2 kurung siku bcs bukan menjadi pihak input

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
# membangun model

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#evaluasi model
y_pred = model.predict(x_test)

mse = mean_squared_error(y_pred, y_test)
print(f"Mean Squared Error {mse}")

mse
```

```python
# visualisasi hasil
plt.scatter(x, y, color= 'blue')
plt.plot(x, model.predict(x), color= 'brown')
plt.title('Linear Regression')
plt.xlabel('Jumlah Iklan')
plt.ylabel('Jumlah Pembeli')
```
