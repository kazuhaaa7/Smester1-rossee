```python
# imporr library yg diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

```

```python
# membuat data

data = {
    "Luas_Kamar" : [12, 15, 18, 20, 22, 25, 30],
    "Harga_Sewa" : [250, 275, 310, 350, 370, 410, 480]
}

datas = data
df = pd.DataFrame(datas)


df
```

```python
#ekspolrasi data(memviusualisasikan data) 
sns.scatterplot(data=df, x= "Luas_Kamar", y= "Harga_Sewa")
plt.title('Harga Luas kamar dengan harga sewa')
plt.show()
```

```python
# persiapan data
a = df[['Luas_Kamar']] #features
b= df['Harga_Sewa'] #target

a_train, a_test, b_train, b_test = train_test_split(a,b, test_size = 0.2, random_state = 42)
```

```python
# membangun model

model = LinearRegression()
model.fit(a_train, b_train)
```

```python
#evaluasi model
b_pred = model.predict(a_test)

mse = mean_squared_error(a_test, b_pred)
print(f"Mean Squared Error {mse}")
```

```python
# visualisasi hasil
plt.scatter(a, b, color= 'blue')
plt.plot(a, model.predict(a), color= 'brown')
plt.title('Linear Regression')
plt.xlabel('Luas kamar')
plt.ylabel('Harga sewa')
plt.show()
```
