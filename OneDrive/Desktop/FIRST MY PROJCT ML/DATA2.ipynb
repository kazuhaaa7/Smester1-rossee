```python
df_gp_rating_mode = df_gp['Rating'].mode()[0]
df_gp_type_mode = df_gp['Type'].mode()[0]
df_gp_cr_mode = df_gp['Content Rating'].mode()[0]
df_gp_cv_mode = df_gp['Current Ver'].mode()[0]
df_gp_av_mode = df_gp['Android Ver'].mode()[0]


df_gp['Rating'].fillna(df_gp_rating_mode, inplace=True)
df_gp['Type'].fillna(df_gp_type_mode, inplace=True)
df_gp['Content Rating'].fillna(df_gp_cr_mode, inplace=True)
df_gp['Current Ver'].fillna(df_gp_cv_mode, inplace=True)
df_gp['Android Ver'].fillna(df_gp_av_mode, inplace=True)
```
